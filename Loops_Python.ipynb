# FOR LOOP


```python
sport_games = ['football', 'football - American', 'hockey', 'baseball', 'cricket']


for games in sport_games: #for <temporary variable> in <list variable>:<action>
  print(games)
```

```python
#Using Range in Loops. This will help us print a string multiple times
promise = "I will not chew gum in class"

for i in range(5):
  print(promise)
```

```python
#Break in For Loop
dog_breeds_available_for_adoption = ['french_bulldog', 'dalmatian', 'shihtzu', 'poodle', 'collie']
dog_breed_I_want = 'dalmatian'
for dogs_available in dog_breeds_available_for_adoption:
    print(dogs_available)
    if dogs_available== dog_breed_I_want:
        print("They have the dog I want!")
        break
print("I have found my Dog Breed")
```
